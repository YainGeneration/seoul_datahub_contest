```python
import pandas as pd
```

### 장소 리스트 정리

```python
parks_df = pd.read_csv("selected_parks.csv")
parks_df = parks_df[["park_name"]]
parks_df["category"] = "주요공원"
parks_df
```

```python
han_river_df = pd.DataFrame({
    "park_name": ['강서한강공원', '난지한강공원', '양화한강공원', '망원한강공원', '여의도한강공원', '이촌한강공원', 
                   '반포한강공원', '잠원한강공원', '뚝섬한강공원', '잠실한강공원', '광나루한강공원'],
    "category": "한강공원"
})
han_river_df
```

```python
places_df = pd.concat([parks_df, han_river_df], ignore_index=True)
places_df
```

### 검색어 조합 생성

```python
search_queries = []
for _, row in places_df.iterrows():
    for suffix in selected_keywords:
        search_queries.append({
            "search_query": f"{row['park_name']} {suffix}",
            "park_name": row["park_name"],
            "category": row["category"],  # 그대로 유지 가능
            "suffix": suffix
        })
```

```python
search_df = pd.DataFrame(search_queries)
search_df
```

### 저장

```python
search_df.to_csv("search_query_list.csv", index=False)
```
